```python
import pandas as pd
```

```python
train = pd.read_csv('train.csv')
target = train['SalePrice']
df = pd.read_csv('processed_train.csv')
data = df.join(target)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(data.values, target.values, test_size = 0.25)
```

```python
from xgboost import XGBRegressor
my_model = XGBRegressor(n_estimators = 1000, learning_rate = 0.05)
my_model.fit(train_X, train_y, early_stopping_rounds = 5, eval_set = [(test_X, test_y)], verbose = False)
test_predict = my_model.predict(test_X)
```

```python
from sklearn.metrics import mean_absolute_error
print('Mean Absolute Error: ', str(mean_absolute_error(test_predict, test_y)))
```
